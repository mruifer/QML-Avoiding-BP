# Looking for BP using the $\hat{Z}_1 \hat{Z}_2$ observable

We start by studying a simple case: the minimization of an expectation value of a simple, two qubits, observable. In order to look for BP, we will study the cocentration of the cost function and its gradient using different number of qubits.

In order to do that, we calculate the variances of the observable and its derivatives. If there is a BP, one should expect to see an exponential reduction of the variances as the number of qubits increase.

First, we import our custom modules that includes functions capable of expand the observable, build different ansatzs and calculate the variances for different numbers of qubits.

```python
import sys
import numpy as np
import matplotlib.pyplot as plt
sys.path.append('../../')

from src import variances as var, ansatzs as anz

from qiskit.quantum_info import SparsePauliOp
```

We define the observable:

```python
observable = SparsePauliOp.from_list([("ZZ",1)])
print(observable)
```

We use our custom deep ansatz:

```python
# set the number of layer per qubit
def ansatz_fun(num_qubits):
    return anz.build_deep_ansatz(num_qubits, 10)
```

Calculate the variances concentration:

```python
data = var.variance_vs_nQubits(ansatz_fun, 2, 15, observable, 0, 100, print_info=True, plot_info=True, do_regress=True)
```

We can see a clearly decreasing linear tendecy. Taking into account that the vertical axis is in logaritmic scale, we conclude that we have found an exponential concentration of the variances, thus, a barren plateau.

## Noise induced BP

We will use out custom function that extracts the noise model from the fake backend provided and escales it given the noise_scale variable.

We study the same concentration for different noise scales:

```python
from qiskit.providers.fake_provider import FakeManila

fake_backend = FakeManila()

data_noise = []

for noise_scale in np.linspace(0, 0.15, 8):
    data_noise.append(var.noisy_variance_vs_nQubits(ansatz_fun, fake_backend, noise_scale, 2, 10, observable, 0, 100, print_info=True, plot_info=True, do_regress=True))
```

### We try using less layers

As the graphs decays very qickly to levels near $10^{-3}$ for expected value and $10^{-4}$ for its derivative, were it doesn't concentrate anymore. We will try to use less layers in the deep ansatz, as well as perform the calculations with lower noise levels.

```python
from qiskit.providers.fake_provider import FakeMontreal

fake_backend = FakeMontreal()

data_noise = []

def low_layers_ansatz_fun(n_qubits):

    return anz.build_deep_ansatz(n_qubits, 2)

for noise_scale in np.linspace(0, 0.05, 6):
    data_noise.append(var.noisy_variance_vs_nQubits(low_layers_ansatz_fun, fake_backend, noise_scale, 2, 9, observable, 0, 100, print_info=True, plot_info=True, do_regress=False))
```

We got some interesting results, as the variances concentrates faster while increasing the noise, without decaying to the minimum suddenly. We consider a noise scale of 0.02 as a good point to represent the differences between tjhe noise free system and the noisy system.

```python
import matplotlib.pyplot as plt
plt.plot(data_noise[0]["n_qubits"], data_noise[0]["var_value"], marker="o", color = "black", label="Libre de ruido")
plt.plot(data_noise[0]["n_qubits"], data_noise[3]["var_value"], marker="o", color = "red", label=r"Ruido del backend Montreal $\times$ 0.02")

plt.xlabel(r"$N$ qubits", size=16)
plt.ylabel(r"Var($\langle O\rangle$)", size=16)
plt.title("NIBP para valor esperado", size=16)
plt.xticks(fontsize="15")
plt.yticks(fontsize="15")

plt.yscale("log")
plt.show()

plt.plot(data_noise[0]["n_qubits"], data_noise[0]["var_deriv"], marker="o", color = "Black", label="Libre de ruido")
plt.plot(data_noise[0]["n_qubits"], data_noise[3]["var_deriv"], marker="o", color = "red", label=r"Ruido del backend Montreal $\times$ 0.02")

plt.xlabel(r"$N$ qubits", size=16)
plt.ylabel(r"Var($\partial_0\langle O\rangle$)", size=16)
plt.title(r"NIBP para derivada respecto $\theta_0$ del valor esperado", size=16)
plt.xticks(fontsize="15")
plt.yticks(fontsize="15")

plt.yscale("log")
plt.legend(fontsize="15")
plt.show()
```
