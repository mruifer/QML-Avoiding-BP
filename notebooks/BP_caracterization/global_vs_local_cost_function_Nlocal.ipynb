## Hamiltoniano H2 con Nlocal

```python
def build_Nlocal(num_qubits):
    ansatz, num_params = anz.build_Nlocal_ansatz(num_qubits, layers=2)
    return ansatz, num_params

with open("data/H2_hamiltonian.pkl", "rb") as f:
    obs = pickle.load(f)

data = []

data = var.variance_vs_nQubits(build_Nlocal, 4, 15, base_observable = obs, index=2, num_shots=1000, do_regress=True)
```

```python
# Datos
x = data["n_qubits"]
y = data["var_value"]

# Transformacin logartmica del eje y
log_y = np.log(y)

# Regresin lineal
slope, intercept, r_value, p_value, std_err = linregress(x, log_y)

# Recta ajustada
x_fit = np.linspace(min(x), max(x), 100)
y_fit = np.exp(slope * x_fit + intercept)

# Plot
plt.scatter(x, y, label=r"Var($\langle O\rangle$)")
plt.xlabel("N qubits", size=16)
plt.yscale("log")
plt.xticks(fontsize="15")
plt.yticks(fontsize="15")
plt.legend(loc="upper right", bbox_to_anchor=(1.0, 1), fontsize="15")
plt.show()

```
